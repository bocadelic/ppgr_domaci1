```python
import numpy as np
```

Koordinate tacaka

```python
p1 = np.array([844, 296])
p2 = np.array([456, 637])
p3 = np.array([161, 438])
p5 = np.array([877, 164])
p6 = np.array([463, 470])
p7 = np.array([133, 284])
p8 = np.array([609, 75])
points = [p1, p2, p3, p5, p6, p7, p8]
```

Funkcija za racunanje osmog temena

```python
def osmoteme(tacke):
    for i in range(len(tacke)):
        tacke[i] = np.append(tacke[i], 1)
        
    xb1 = np.cross(np.cross(tacke[1], tacke[4]), np.cross(tacke[0], tacke[3]))
    xb2 = np.cross(np.cross(tacke[1], tacke[4]), np.cross(tacke[5], tacke[2]))
    xb3 = np.cross(np.cross(tacke[5], tacke[2]), np.cross(tacke[3], tacke[0]))
    
    yb1 = np.cross(np.cross(tacke[4], tacke[3]), np.cross(tacke[5], tacke[6]))
    yb2 = np.cross(np.cross(tacke[1], tacke[0]), np.cross(tacke[4], tacke[3]))
    yb3 = np.cross(np.cross(tacke[5], tacke[6]), np.cross(tacke[1], tacke[0]))
    
    zb1 = np.cross(np.cross(tacke[1], tacke[2]), np.cross(tacke[4], tacke[5]))
    zb2 = np.cross(np.cross(tacke[3], tacke[6]), np.cross(tacke[4], tacke[5]))
    zb3 = np.cross(np.cross(tacke[1], tacke[2]), np.cross(tacke[3], tacke[6]))
    
    xb = (xb1 + xb2 + xb3) / 3
    yb = (yb1 + yb2 + yb3) / 3
    zb = (zb1 + zb2 + zb3) / 3
    
    p41 = np.cross(np.cross(xb, tacke[6]), np.cross(yb, tacke[2]))
    p42 = np.cross(np.cross(xb, tacke[6]), np.cross(zb, tacke[0]))
    p43 = np.cross(np.cross(yb, tacke[2]), np.cross(zb, tacke[0]))
    
    p4 = (p41 + p42 + p43) / 3
    rezultat = np.array([np.round(p4[0] / p4[2]), np.round(p4[1] / p4[2])])
        
    return rezultat
    
```

Osmo teme

```python
p4 = osmoteme(points)
print(p4)
```
